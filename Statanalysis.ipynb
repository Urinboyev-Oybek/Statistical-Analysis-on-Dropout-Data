## Do actual dropout rates amount genders are same or females' dropout rates are higher

1) **Hypothesis testing**:

H0: population_mean_of_female = population_mean_of_male

H1: population_mean_of_female < population_mean_of_male

2) significant level = 0.05
   **Decision rule**
   IF testing point <critical point, reject H0 esle accept H0.


```python
data_male, data_female = df[df['Gender']=='Male'], df[df['Gender'].isin(['Female'])]
df['Gender'].value_counts() 
```

Sample size for females is 5011, sample size for males is 4989

```python
nm = len(data_male) 
nf = len(data_female)
print(f" Sample size of male: {nm}")
print(f" Sample size of female: {nf}")
```

```python
sm = (data_male['Dropout']==1).sum() /nm
sf = (data_male['Dropout']==1).sum()/nf
print(f" Sample male proportion: {pm:.2f}")
print(f" Sample female proportion: {pf:.2f}")
```

```python
pp = (nm *pm + nf * pf)/(nm+nf)
qp = 1-pp
print(f"Pooled proportion: {pp:.2f}")
```

```python
critical_point = 0-1.645*(pp*qp/nm +pp*qp/nf)**0.5
print(f"Critical point of this study: {critical_point:.2f}")
```

```python
testing_point = sf-sm
print(f"Testing point of this study is {testing_point:.3f}")
```

**Based on the hypothesis test, there is no statistically significant evidence to suggest a difference in dropout rates between males and females. Therefore, we fail to reject the null hypothesis.**

## DO ACTUAL DROPOUT RATES ARE SAME AMOUNG SCHOLARSHIP TYPES, OR THE DROPOUT RATE OF STUDENTS WHO HAVE SCHOLARSHIP IS LOWER.

**Hypothesis testing**

H0: population_mean_scholarship_yes = population_mean_scholarship_no

H1: population_mean_scholarship_yes < population_mean_scholarship_no

**Decision Rule** at significant level is 0.05

If testing_point<critical_point, reject H0 else accept H0

```python
dy = df[df['Scholarship']=='Yes']
dn = df[df['Scholarship']=='No']
```

```python
ny = len(dy)
nn = len(dn)
```

```python
sy = (dy['Dropout']==1).sum()/ny
sn = (dn['Dropout']==1).sum()/nn
print(f"Sample proportion where Scholarship is Yes: {sy:.3f}")
print(f"Sample proportion where Scholarship is No : {sn:.3f}")
```

```python
pp_s = (ny*sy+nn*sn)/(ny+nn)
qp_s=1-pp_s
print(f"Pooled proportion: {pp:.3f}")
```

```python
testing_point = sy-sn
critical_point =0-1.645* (pp_s*qp_s/ny +pp_s*qp_s/nn)**0.5
print(f"Testing point of this study : {testing_point:.3f}")
print(f"Critical point of this study: {critical_point:.3f}")
```

```python
if testing_point < critical_point:
    print("Reject H0")
else:
    print('Accept H0')
```

**Based on the hypothesis test, there is no statistically significant evidence to suggest a difference in dropout rates between scholarship types. Therefore, we fail to reject the null hypothesis.**
