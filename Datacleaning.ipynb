```python
import pandas as pd
```

```python
df = pd.read_csv('student_dropout_dataset_v3.csv')
```

```python
df.head()
```

```python
df.drop_duplicates()
```

```python
df.describe()
```

```python
df.isnull().mean() * 100
```

```python
# As we can see, the percentage of missing values in some columns are almost all less than or equal to 5%.  
# Therefore, filling them with median or mode is more better way to keep all valuable data rather than just droppig
# the rows
```

```python
df['Parental_Education'] = df['Parental_Education'].fillna('Unknown')
```

```python
df['Parental_Education'].value_counts()
```

```python
# I filled none values in Parental Education with 'Unknown' since this column only stored categorical data.
```

```python
fam_inc_med = df.groupby('Parental_Education')['Family_Income'].transform('median')
df['Family_Income'] = df['Family_Income'].fillna(fam_inc_med)
```

```python
df['Family_Income'].isnull().sum()
```

```python
# I filled the missing values in Family income using meadians of each subgroups within Parental Education.
#  As result, we have appropraite data since the income is positive corelated to education level in more cases.
# Using median helps us fill the null values without any high-income outliers' effect
```

```python
sts_inx_rate = df.groupby(['Part_Time_Job','Scholarship'])['Stress_Index'].transform('median')
```

```python
df['Stress_Index'] = df['Stress_Index'].fillna(sts_inx_rate)
```

```python
df['Stress_Index'].isnull().sum()
```

```python
df[['Scholarship' ,'Part_Time_Job', 'Stress_Index']]
```

```python
df['Stress_Index'].isnull().sum()
```

```python
# Filled missing Stress_Index values using median scores grouped by Scholarship and Part_Time_Job
# status to get more appropriate, context-specific estimates.
```

```python
df.isnull().mean()*100
```

```python
df['Study_Hours_per_Day'] = pd.to_numeric(df['Study_Hours_per_Day'] , errors='coerce' )
study_hour_pd = df.groupby(['Gender', 'Internet_Access', 'Part_Time_Job', 'Scholarship', 'Department'])['Study_Hours_per_Day'].transform('median')
df['Study_Hours_per_Day']=df['Study_Hours_per_Day'].fillna(study_hour_pd)
```

```python
# Filled missing Study_Hours_per_Day values using median scores grouped by Gender, Internet_Access, Scholarship, Part_Time_Job and Department
# status to get more appropriate, context-specific estimates.
# I believe that these subgroups can help us to estimate with more accuracy
```

```python
df.to_csv('student_drop_out_new_dataset.csv', index= False)
```
